```python
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
```

## Data

```python
dir_path = "../process/"
#logs_df = pd.read_csv(f"{dir_path}external_test_data.csv", encoding='utf-8')
logs_df_1 = pd.read_csv(f"{dir_path}3_22external_test_data_heuristic_prediction.csv", encoding='utf-8')
print(logs_df_1.shape)
logs_df_1.head()
```

```python
logs_df_2 = pd.read_csv(f"{dir_path}4_10external_test_data_tdidf_prediction.csv", encoding='utf-8')
print(logs_df_2.shape)
logs_df_2.head()
```

```python
logs_df_3= pd.read_csv(f"{dir_path}5_10external_test_data_word2vect_prediction.csv", encoding='utf-8')
print(logs_df_3.shape)
logs_df_3.head()
```

## Evaluacion global 

```python
# Reporte general para todo el dataset
print("EH - Evaluación global:")
print(classification_report(logs_df_1['Level'], logs_df_1['PrioridadEtiqueta'], digits=4))

# También puedes imprimir matriz de confusión si quieres verla
print("EH - Matriz de confusión:")
print(confusion_matrix(logs_df_1['Level'], logs_df_1['PrioridadEtiqueta']))
```

```python
# Reporte general para todo el dataset
print("LR - Evaluación global:")
print(classification_report(logs_df_2['Level'], logs_df_2['Predicted_Level'], digits=4))

# También puedes imprimir matriz de confusión si quieres verla
print("LR - Matriz de confusión:")
print(confusion_matrix(logs_df_2['Level'], logs_df_2['Predicted_Level']))
```

```python
# Reporte general para todo el dataset
print("LR - Evaluación global:")
print(classification_report(logs_df_3['Level'], logs_df_3['Predicted_Level'], digits=4))

# También puedes imprimir matriz de confusión si quieres verla
print("LR - Matriz de confusión:")
print(confusion_matrix(logs_df_3['Level'], logs_df_3['Predicted_Level']))
```

## Evaluacion por source

```python
sources = logs_df_1['source'].unique()
sources
```

## **EH**

```python
sources = logs_df_1['source'].unique()

for src in sources:
    print(f"\nEvaluación para source: {src}")
    df_src = logs_df_1[logs_df_1['source'] == src]
    print(classification_report(df_src['Level'], df_src['PrioridadEtiqueta'], digits=4))
    labels = sorted(df_src['Level'].unique())
    print("Orden de categorías:", labels)
    
    cm = confusion_matrix(df_src['Level'], df_src['PrioridadEtiqueta'], labels=labels)
    print(cm)
```

## **LR y TF-IDF**

```python
sources = logs_df_2['source'].unique()

for src in sources:
    print(f"\nEvaluación para source: {src}")
    df_src = logs_df_2[logs_df_2['source'] == src]
    print(classification_report(df_src['Level'], df_src['Predicted_Level'], digits=4))
    labels = sorted(df_src['Level'].unique())
    print("Orden de categorías:", labels)
    
    cm = confusion_matrix(df_src['Level'], df_src['Predicted_Level'], labels=labels)
    print(cm)
```
